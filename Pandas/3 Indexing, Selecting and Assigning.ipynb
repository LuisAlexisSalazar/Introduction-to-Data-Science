# Indexing
**Selección particular de filas y columnas** <br>
**loc: Label based selection** <br>
**iloc: Index based selection**

## Index Based Selection (iloc)
**Usado para la selección de datos basados en la posición numerica** <br>
**Sintaxis es df.iloc[]**

```python
import pandas as pd
exam_scores = pd.read_csv('data/exam_scores.csv')
exam_scores.head()
```

```python
# Select first row and first column
exam_scores.iloc[0, 0]
```

```python
exam_scores.iloc[0:5, 4]
```

```python
# Si no se pone nada selecciona todo los elementos exam_scores.iloc[:,:]
```

```python
# Pasar lista de indices para determinar que columnas usar y el 2° parametro es que valor retornar de las filas
list_indexs = [0,1,2,3]
# colums and values
exam_scores.iloc[list_indexs,2]
```

```python
# Using negative index
exam_scores.iloc[-5:, 0:2]
```

## Label Based Selection (loc)
**Basado en el nombre/indice de la fila o la columna** <br>
**loc e iloc son conceptualmente similares. La diferencia es que iloc considera la indexación predeterminada mientras que loc ignora la indexación predeterminada.** <br>
**loc se utiliza para seleccionar datos en función del valor/nombre del índice de datos, no de las posiciones numéricas.**

```python
exam_scores.loc[0,'race/ethnicity']
```

```python
exam_scores.loc[0:5,['gender','lunch','math score']]
```
