<h1>Index with semicolon (:)</h1>

```python
num_arr
```

```python
# [0, 0:2] will get you the row indexed at 0 and 
#the columns indexed from 0 (inclusive) to 2 (exclusive).
num_arr[0,0:2]
```

```python
num_arr = np.array(np.arange(16))
num_arr = num_arr.reshape(4,4)
num_arr
```

```python
# Index Row 1 since 3 (exclusive) -> 1->2
# Index 2 since 4 (exclusive) -> 2-3

num_arr[1:3, 2:4]
```

```python
arr = np.arange(start = 10,stop = 50, step=10)
arr
```

```python
arr[-2]
```

```python
arr = np.arange(2,13,2)
arr = arr.reshape(2,3)
arr
```

```python
arr[-2]
```

<h1>Negative Index Interval<h/1>

```python
arr = np.arange(0,16).reshape(4,4)
arr[1:3,2:4]
```

```python
arr = np.arange(0,16).reshape(4,4)
arr[-3:-1,-2:]
```

<h1>Striding ->  [Star_index: End_index: Stride]<h1>

```python
# Striding : The steps to be taken while moving in a direction
# stride -> the step you want to jump
arr = np.arange(0,16).reshape(4,4)
arr
```

```python
# 1st and 2nd are index start and end index (exlcusive)
# The third parameter specifies the stride
arr[0:3:2]
```

```python
# Other Example with string
ss = "Sammy Shark!"
```

```python
print(ss[6:11:1])
```

```python
# Each pass 2 of step i.e S->M->Y  omited S "A" M "M" Y
print(ss[0:12:2])
```
